```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

from eden.util import configure_logging
import logging


DEBUG=True
NJOBS=4
if DEBUG: NJOBS=1

configure_logging(logging.getLogger(),verbosity=1+DEBUG)
from IPython.core.display import HTML
HTML('<style>.container { width:95% !important; }</style>')

# data source, see introduction for info.
from eden.converter.graph.gspan import gspan_to_eden
from itertools import islice
def get_graphs(dataset_fname='../toolsdata/bursi.pos.gspan', size=100):
    return islice(gspan_to_eden(dataset_fname),size)


#import warnings
#warnings.filterwarnings('ignore')
```

```python
%%time
import graphlearn.minor.cascade as cascade
from graphlearn.utils import draw

graphs=get_graphs(size=200)
cascade=cascade.cascade(depth=5)
cascade.setup_transformers()
minor_to_nested = lambda x: cascade.decomposer.make_new_decomposer(x).pre_vectorizer_graph(nested=True)
graphs=cascade.write_layer_to_graphs(graphs,0)
```

```python
%%time
DEPTH=0
graphs = [minor for minor in cascade.transformers[DEPTH].fit_transform(graphs)]
graphs = map( minor_to_nested,graphs)
draw.graphlearn_layered(graphs[:4], vertex_label='label')
draw.graphlearn_dict(cascade.transformers[DEPTH].graphclusters)

DEPTH=1
graphs = [minor for minor in cascade.transformers[DEPTH].fit_transform(graphs)]
graphs = map( minor_to_nested,graphs)
draw.graphlearn_layered(graphs[:5], vertex_label='label')
draw.graphlearn_dict(cascade.transformers[DEPTH].graphclusters)

DEPTH=2
graphs = [minor for minor in cascade.transformers[DEPTH].fit_transform(graphs)]
graphs = map( minor_to_nested,graphs)
draw.graphlearn_layered(graphs[:5], vertex_label='label')
draw.graphlearn_dict(cascade.transformers[DEPTH].graphclusters)

DEPTH=3
graphs = [minor for minor in cascade.transformers[DEPTH].fit_transform(graphs)]
graphs = map( minor_to_nested,graphs)
draw.graphlearn_layered(graphs[:5], vertex_label='label')
draw.graphlearn_dict(cascade.transformers[DEPTH].graphclusters)
```
